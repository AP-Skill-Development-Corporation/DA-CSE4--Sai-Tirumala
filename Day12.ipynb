```python
x=[4,5,6,8,7]
y=[.4,3.5,4.6,7.8,10.3]
plt.plot(x,y)
plt.title('Basic Graph')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
z=[34.5,67.8,56.7,34.5,23.6]
plt.plot(x,y) # plotting sample line graph b/w x&y
plt.xlabel('x-axis') # adding label on the x-axis
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(y,z)
```

```python
#while plotting the multiple plot in a plain
plt.plot(x,y,label='first')
plt.plot(x,z,label="second")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc=2) # upper right by default unless you pass loc=value
plt.title("Second Graph")
plt.show()
```

```python
plt.plot(x,y,linestyle='--',color='red')
plt.show()
```

```python
classes=['Python','DA','R','AI','ML']
st1=[10,40,30,40,20]
st2=[50,30,27,15,60]
st3=[20,15,5,30,23]
plt.bar(classes,st1)
```

```python
from matplotlib import style
style.use('ggplot')
plt.bar(classes,st2)
```

```python
plt.bar(classes,st2,color='black')
```

```python
#attributes to the bar graph
plt.bar(classes,st3,width=0.5,color='y',align='edge',edgecolor='m',linestyle='-.',linewidth=0.8)
```

```python
#multiple bar graphs in a plain
plt.bar(classes,st1)
plt.bar(classes,st2)
plt.bar(classes,st3)
```

```python
#scatter plot
#scattering the points
x=np.linspace(1,10,40)
y=np.sin(x)
plt.scatter(x,y,color='g',s=40) # s means size of the point/marker
plt.show()
```

```python
#histogram
#plt.hist
one=np.random.randint(10,40,100)
two=np.random.randint(15,40,100)
plt.hist(one,color='b',bins=50)# for 100 data points,50 bins representation
plt.xlabel('Age category')
plt.ylabel('Voters')
plt.show()
```

```python
plt.hist(two,bins=20,color='g',orientation='horizontal')
plt.show()
```

```python
plt.hist(one,color='y',bins=40)
plt.hist(two,color='g',bins=60)
plt.title('Multi Histograms')
plt.show()
```

```python
# pie plot
#making the regions
x=[56,25,30,24]
plt.pie(x)
plt.show()
```

```python
plt.pie(rn)
plt.show()
```

```python
plt.pie(sec)
plt.show()
```

```python
plt.pie(x,labels=['Good','Average','Above average','Excellent'])
plt.show()
```

```python
fbk=['Good','Average','Above average','Excellent']
plt.pie(x,colors=['green','cyan','blue','pink'],labels=fbk)
plt.show()
```

```python
fbk=['Good','Average','Above average','Excellent']
plt.pie(x,colors=['green','cyan','blue','pink'],labels=fbk,autopct='%f.%%')
plt.title('Sort by Responses')
plt.show()
```

```python
#exclude some part from the plot
exc=[0,0,0,1]
plt.pie(x,labels=fbk,colors=['green','cyan','blue','pink'],autopct='%f.%%',
        shadow=True,startangle=180,explode=exc)
plt.title('Excluded')
plt.show()
```

```python
#one by one arragement
#scatter plot
a=[9,3,5,4,8]
b=[9,4,5,6,2]
c=[6,8,2,3,1]
x=np.array([8,4,9,4,2])
plt.stackplot(x,a,b,c)
plt.show()
```

```python
f=[1,2,3,4,5]
s={'y1':[5,6,4,5,7],'y2':[1,6,4,5,6],
  'y3':[1,1,2,3,4]}
plt.stackplot(f,s.values())
plt.show()
```
